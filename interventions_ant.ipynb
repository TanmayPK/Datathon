```python
interventions1 = pd.read_parquet('data/interventions1.parquet')
interventions2 = pd.read_parquet('data/interventions2.parquet')
interventions3 = pd.read_parquet('data/interventions3.parquet')
```

```python
interven_combined = pd.concat([interventions1, interventions2, interventions3])
```

```python
antwerpen_interventions = interven_combined[interven_combined['Province intervention'].str.contains('ANT')]
```

```python
import math
antwerpen_interventions['Latitude intervention'] = antwerpen_interventions['Latitude intervention'].apply(lambda x: x / 10 if x >= 100 else x)
antwerpen_interventions['Longitude intervention'] = antwerpen_interventions['Longitude intervention'].apply(lambda x: x / 100 if x >= 100 else x)
```
