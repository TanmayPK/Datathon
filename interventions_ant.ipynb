```python
import pandas as pd
```

```python
interventions1 = pd.read_parquet('data/interventions1.parquet')
interventions2 = pd.read_parquet('data/interventions2.parquet')
interventions3 = pd.read_parquet('data/interventions3.parquet')
```

```python
interventions1.head(5)
```

```python
interventions1.iloc[0]
```

```python
interventions1.columns
```

```python
interven_combined = pd.concat([interventions1, interventions2, interventions3])
```

```python
antwerpen_interventions = interven_combined[interven_combined['Province intervention'].str.contains('ANT')]
```

```python
antwerpen_interventions
```

```python
duplicate_count = antwerpen_interventions.duplicated().sum()
duplicate_count
```

```python
import math
antwerpen_interventions['Latitude intervention'] = antwerpen_interventions['Latitude intervention'].apply(lambda x: x / 10 if x >= 100 else x)
antwerpen_interventions['Longitude intervention'] = antwerpen_interventions['Longitude intervention'].apply(lambda x: x / 100 if x >= 100 else x)
```

```python
min_latitude = antwerpen_interventions['Latitude intervention'].min()
max_latitude = antwerpen_interventions['Latitude intervention'].max()

min_longitude = antwerpen_interventions['Longitude intervention'].min()
max_longitude = antwerpen_interventions['Longitude intervention'].max()

min_latitude, max_latitude, min_longitude, max_longitude
```

```python
antwerpen_interventions['Longitude intervention'][antwerpen_interventions['Longitude intervention']> 10] /=10
```
