```python
from geopy import Nominatim
import pandas as pd
aed = pd.read_parquet('data/aed_locations.parquet.gzip')
```

```python
print(aed["address"])
addresses = aed["address"].tolist()

for address in addresses:
    address = address.split(" ")
    for word in address:
        if len(word) != 0:
            if word[-1] == ".":
                print(word)
```

```python
locator = Nominatim(user_agent="ouiii")
location = locator.geocode("4630, Ayeneux, Liège, Belgique")
print(location)
```

```python
print("Latitude = {}, Longitude = {}".format(location.latitude, location.longitude))
```
