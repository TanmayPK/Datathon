```python
import pandas as pd
```

```python
aed = pd.read_parquet('data/aed_locations.parquet.gzip')
```

```python
aed['province'].value_counts()
```

```python
antwerpen = aed[aed['province'] == 'Antwerpen']
```

```python
antwerpen
```

```python
antwerpen[antwerpen['address'].str.contains('\.')]
```

```python
antwerpen['concat_address'] = antwerpen['address'] + ' ' + antwerpen['number'].fillna('').astype(str).apply(lambda x: str(int(float(x))) if x.strip() and '.' in x else x) + ', ' + antwerpen['province'] + ', ' + antwerpen['postal_code'].fillna('').astype(str).apply(lambda x: str(int(float(x))) if x.strip() and '.' in x else x) + ', Belgique'
```

```python
antwerpen['concat_address']
```

```python
from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
```

```python
locator = Nominatim(user_agent="wigsyo")
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=0.1)
```

```python
tqdm.pandas(desc="geolocation")
antwerpen['location_geo'] = antwerpen['concat_address'].apply(geocode)
```

```python
antwerpen['coords'] = antwerpen['location_geo'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
antwerpen[antwerpen['coords'].isna()]
```

```python
antwerpen[['lat', 'lon', 'alt']] = pd.DataFrame(antwerpen['coords'].tolist(), index=antwerpen.index)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
BBox = ((antwerpen.lon.min(), antwerpen.lon.max(), antwerpen.lat.min(), antwerpen.lat.max()))
```

```python
BBox
```

```python
ant_map = plt.imread('img/ant.png')
```

```python
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(antwerpen.lon, antwerpen.lat, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('AED Locations - Antwerpen')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ant_map, zorder=0, extent = BBox, aspect= 'equal')
```
